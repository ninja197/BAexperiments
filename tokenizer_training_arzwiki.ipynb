```python
from datasets import load_dataset

dataset = load_dataset("wikipedia", language="arz", date="20220401", beam_runner="DirectRunner")
```

```python
dataset
```

```python
# iterator -> only load in memory when requested
batch_size = 1000
def batch_iterator():
    for i in range(0, len(dataset), batch_size):
        yield dataset[i : i + batch_size]["text"]
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")
```

```python
tokenizer.is_fast
```

```python
new_tokenizer = tokenizer.train_new_from_iterator(batch_iterator(), vocab_size=30000)
```

```python
new_tokenizer.save_pretrained("arzwiki")
```
